```python
#Sort by Market Cap - descending so we can isolate the top 25
sorted_asset_data_df = asset_data_df.sort_values('CAPT', ascending = False)
sorted_asset_data_df.head()
```

```python
#Keep only top 25 rows and set indexes so we only have numeric columns
top_25_cap = sorted_asset_data_df.iloc[:25, :]
top_25_cap = top_25_cap.set_index('TICKER')
top_25_cap = top_25_cap.drop('NAME', axis = 1)
top_25_cap
```

```python
#Store portfolio weights by market cap to be used later
pf_weights_by_cap = top_25_cap['CAPT'] / np.sum(top_25_cap['CAPT'])
pf_weights_by_cap = dict(pf_weights_by_cap)
pf_weights_by_cap
```

#### a)

```python
b = np.array([.02, .01, .03])
b = b.reshape([3, 1])
b = b.transpose()
b.shape
```

```python
def find_alpha(ticker, top_25_df, b):
    
    #b is 1x3
    #X matrix is 3x1
    
    #Filter to ticker
    filtered_df = top_25_df[top_25_df.index == ticker]
    
    earnyld_col = np.where(filtered_df.columns == 'EARNYLD')[0][0]
    value_col = np.where(filtered_df.columns == 'VALUE')[0][0]
    momentum_col = np.where(filtered_df.columns == 'MOMENTUM')[0][0]
    
    ticker_earnyld = filtered_df.iloc[:, earnyld_col]
    ticker_value = filtered_df.iloc[:, value_col]
    momentum_col = filtered_df.iloc[:, momentum_col]
    
    X = np.array([ticker_earnyld,
                  ticker_value,
                  momentum_col])
    
    return (b @ X)[0][0]
```

```python
#Calculate alpha for all top 25 stocks
alpha_dict = {ticker : find_alpha(ticker, top_25_cap, b) for ticker in top_25_cap.index}
alpha_dict
```

```python
alphas_sorted = sorted(alpha_dict.items(), key=lambda x: x[1], reverse=True)
print(str(alphas_sorted[0][0]) + ' has the highest alpha of: ' + str(alphas_sorted[0][1]))
print(str(alphas_sorted[-1][0]) + ' has the lowest alpha of: ' + str(alphas_sorted[-1][1]))
```

```python
large_cap_df = pd.read_csv('Homework 2 Data.csv')
large_cap_df = large_cap_df.drop('Company Name', axis = 1)
large_cap_df = large_cap_df.set_index('Ticker')
large_cap_df['BinaryAlpha'] = np.where(large_cap_df['Broker Rating'] == 'BUY', 1.0, -1.0)
large_cap_df['Alpha_Q8'] = (0.04 * large_cap_df['BinaryAlpha'] * large_cap_df['Active Risk'])
large_cap_df
```

```python
def optimal_holdings(lambd, ticker, df):
    
    filtered_df = df[df.index == ticker]

    alpha = float(filtered_df['BinaryAlpha'])
    omega_squared = float(filtered_df['Active Risk'] ** 2)
    holding = (1 / (2 * lambd)) * (alpha / omega_squared)
    
    return holding
```

```python
lambd = 0.125

holding_dict = {key : optimal_holdings(lambd, key, large_cap_df) for key in large_cap_df.index}
holding_dict
```

```python
holding_sorted = sorted(holding_dict.items(), key=lambda x: x[1], reverse=True)
holding_sorted
```

```python
print('The max holding is ' + str(holding_sorted[0][0] + ' with a weight of ' + str(holding_sorted[0][1])))
```

```python
active_risk_dict = {key : float(large_cap_df[large_cap_df.index == key]['Active Risk']) for key in holding_dict.keys()}
active_risk_dict
```

```python
bet_dict = {key : (holding_dict[key] * active_risk_dict[key]) for key in holding_dict.keys()}
bet_dict
```

```python
bet_sorted = sorted(bet_dict.items(), key=lambda x: x[1], reverse=True)
bet_sorted
```

```python
print('The max bet is ' + str(bet_sorted[0][0] + ' with a bet of ' + str(bet_sorted[0][1])))
```
